```python
import pandas as pd
from os import path
import os

dataPath = path.join(path.dirname(path.dirname(os.getcwd())),
                        'data')
```

### 3.0.1
Load the 2018 aggregated pitching data `(./data/2018-season/pitches.csv)` into a DataFrame named `dfp`. You'll use it for the rest of the problems in this section.

```python
dfp = pd.read_csv(path.join(dataPath, '2018-season/pitches.csv'))
dfp.head()
```

### 3.0.2
Use the dfp DataFrame to create another DataFrame, `dfp50`, that is the top 50 (by lowest ERA) players.

```python
dfp50 = dfp.sort_values('ERA').reset_index().head(50)
```

### 3.0.3
Sort `dfp` by first name in descending order (so Zach Grinke is on the first line). On another line, look at dfp in the repl and make sure it worked.

```python
dfp = dfp.sort_values('nameFirst', ascending=(False))
dfp.head()
```

### 3.0.5
a) Make a new DataFrame, `dfp_simple`, with just the columns `'nameFirst', 'nameLast', 'W', 'L', 'ERA'` in that order.

```python
dfp_simple = (dfp[['nameFirst', 'nameLast',
                   'W', 'L', 'ERA']].copy().reset_index(drop=True))
dfp_simple.head()
```

b) Rarrange `dfp_simple` so the order is `'nameLast', 'nameFirst', 'ERA', 'W', 'L'`.

```python
dfp_simple = dfp_simple[['nameLast', 'nameFirst', 'ERA', 'W', 'L']].copy()
dfp_simple.head()
```

c) Using the original `dfp` DataFrame, add the `'teamID'` column to `dfp_simple`.

```python
dfp_simple['teamID'] = dfp['teamID']
dfp_simple.head()
```

d) Write a copy of `dfp_simple` to `./data/problems/dfp_simple.txt` Make it `'|'` (pipe) delimited instead of `','` (comma) delimited.

```python
dfp_simple.to_csv(path.join(dataPath, 'problems', 'dfp_simple.txt'),
                  sep = '|')
```
